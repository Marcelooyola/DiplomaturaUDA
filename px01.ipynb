```python
df = px.data.gapminder().query("year == 2007")
df
```

*El indice GDP es el producto bruto interno, y en el DataFrame se destacan:*

- lifeExp, como el promedio de la expectativa de vida de las personas.
- pop, como la población del país.
- iso_alpha, el código para gráficos de mapas.

```python
px.strip(df,
         x = "lifeExp")
```

```python
px.strip(df,
         x = "lifeExp",
         hover_name = "country", # hover_name coloca un dato en el tooltip del grafico
         color = "continent", # color permite separar x colores segun determinada agrupación
         template = "gridon") # template gridon permite un formato con fondo transparente y lineas de separación
```

##Histograma

```python
px.histogram(df,
             x = "lifeExp",
             y = "pop",
             template = "gridon",
             hover_name = "country",
             color = "continent"
)
```

Usando Marginal mostrar como estan distribuidos los paises

```python
px.histogram(df,
             x = "lifeExp",
             y = "pop",
             template = "gridon",
             hover_name = "country",
             color = "continent",
             marginal = "rug"
)
```

Separar los continentes en diferentes columnas

```python
px.histogram(df,
             x = "lifeExp",
             y = "pop",
             template = "gridon",
             hover_name = "country",
             color = "continent",
             facet_col= "continent"
)
```

##Grafico de barras

```python
px.bar(df,
       x = "pop",
       y = "continent",
       color = "lifeExp",
       hover_name = "country"
)
```

## Grafico de tortas

Realizar un grafico de tortas donde muestre de dentro hacia afuera Continente - pais y coloree por expectativa de vida

```python
px.sunburst(df,
            values = "pop",
            path = ["continent","country"],
            color = "lifeExp",
            hover_name = "country",
            height = 500,
            template = "presentation"
)
```

##Treemap / mapa de arbol

Realizar un treemap / mapa de arbol donde muestre de englobado [Continente - pais ] y coloree por expectativa de vida

```python
px.treemap(df,
           values = "pop",
           path= ["continent", "country"],
           color = "lifeExp",
           hover_name = "country",
           template = "seaborn",
           height = 500
)
```

##Choropleth

```python
px.choropleth(df,
              color = "lifeExp",
              locations = "iso_alpha",
              hover_name = "country",
              height = 500,
              template = "simple_white")
```

## Scatter
Grafico de dispersión

```python
px.scatter(df,
           x = "gdpPercap",
           y = "lifeExp",
           hover_name = "country",
           template = "simple_white"
)
```

```python
px.scatter(df,
           x = "gdpPercap",
           y = "lifeExp",
           hover_name = "country",
           template = "simple_white",
           color = "continent",
           size = "pop",
           size_max=50
)
```
