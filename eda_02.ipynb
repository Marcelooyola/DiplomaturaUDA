### **Eliminando los NaN**

```
empleados.dropna(how='any', axis=1) # NO HACER, BORRA TODA COLUMNA QUE TENGA AL MEN0S 1 VALOR FALTANTE
```

```
empleados.dropna(how="any", inplace=True) #BORRA TODA FILA QUE TENGA AL MENOS UN VALOR FALTANTE
```

```
empleados
```

### **Completando los NaN**

```
empleados = pd.read_csv(url2)
empleados
```

```
#rellena con el valor anterior a la celda vacia
empleados.edad = empleados.edad.ffill()
empleados
```

```
#rellena con el valor posterior a la celda vacia
empleados.sexo = empleados.sexo.bfill()
empleados
```

```
print(empleados.edad.mean()) #tomar el valor media de la edad
print(empleados.edad.median()) #tomar el valor mediana de la edad
```

```
#relleno los valores de edad faltantes con la media de edad de los valores existentes
empleados["edad"] = empleados.edad.fillna(empleados.edad.mean())
empleados
```

```
empleados.edad = empleados.edad.fillna(empleados.edad.median()) #mediana
empleados.sueldo = round(empleados.sueldo.fillna(empleados.sueldo.mean()),2) #media
empleados.sexo = empleados.sexo.fillna(empleados.sexo.mode()[0]) #moda
empleados
```

```
empleados = pd.read_csv(url2)
#relleno los valores de edad faltantes con la media de edad de los valores existentes y redondeo sin comas
empleados["edad"] = round(empleados.edad.fillna(empleados.edad.mean()),0)
empleados
```

```
empleados = pd.read_csv(url2)
empleados["edad"] = empleados.edad.fillna(empleados.edad.median())
empleados
```

```
empleados["sueldo"] = empleados.sueldo.fillna(empleados.sueldo.median())
empleados
```

## **Transformación de las características de los datos**

```
empleados = pd.read_csv(url2)
```

```
def grado_performance(performance):
    if performance >= 700:
        return "A"
    elif performance < 700 and performance >= 500:
        return "B"
    else:
        return "C"
```

```
empleados["grado_performance"] = empleados.performance.apply(grado_performance)
```

```
empleados
```
